# Clean up the CSV file.
Let us follow the following steps to clean up the csv file.

1) Check the shape of the csv file.
2) Understand the data.
3) Check for any NaN values.
4) Remove NaN values.
5) Remove duplicates.
6) If column names have any spaces, replace with underscore.

```python
import pandas as pd 

df = pd.read_csv('../data/kc_house_data.csv')
df.shape
```

```python
df.sample(10)
```

```python
print(df.isna().sum())
```

```python
print(df.dtypes)
```

```python
df.duplicated().sum()
```

```python
df = df.drop(["id", "date"], axis = 1)
```

```python
df.sample(10)
```

```python
df[df['bedrooms'] > 10]
```

```python
df = df[df['bedrooms'] <= 10]
```

```python
df[df['bathrooms'] == 0]
```

```python
df = df[df['bathrooms'] > 0]
```

```python
df[df['sqft_lot'] > 100000]
```

```python
import numpy as np
df['sqft_lot'] = np.log1p(df['sqft_lot'])
```

```python
df.to_csv('../data/cleaned_house_price.csv', index = False)
```
