# Visualize the data

Let us now try to understand our data, capture the trends and check if there is any need for normalization.

```python
# Price vs date

import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("../data/cleaned_house_price.csv")
```

```python
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.scatterplot(x='sqft_living', y='price', data=df)
plt.title("Price vs Sqft Living")
plt.show()
```

We can observe that the price is increasing as sqft_living increases.

```python
sns.scatterplot(x='bathrooms', y='price', data=df)
plt.title("Price vs Bathrooms")
plt.show()
```

```python
sns.scatterplot(x='bedrooms', y='price', data=df)
plt.title("Price vs Bedrooms")
plt.show()
```

More bedroom does not always indicate higher price.

```python
sns.boxplot(x='grade', y='price', data=df)
plt.title("Price by Grade")
plt.show()
```

Higher the grade, higher the price

```python
sns.boxplot(x='condition', y='price', data=df)
plt.title("Price by Condition")
plt.show()
```

Better condition does not always indicate higher price.

```python
sns.boxplot(x='floors', y='price', data=df)
plt.title("Price by Number of Floors")
plt.show()
```

More floors also does not mean higher price.

```python
plt.figure(figsize=(10, 8))

sns.kdeplot(
    x=df['long'], 
    y=df['lat'], 
    cmap='Blues', 
    fill=True, 
    bw_adjust=0.5, 
    alpha=0.5, 
    cbar=True, 
    weights=df['price']  # Weights by the price to show price intensity
)

plt.title("Density of House Prices by Location")
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```

We can observe from this that some locations have a higher price in general than the others.
